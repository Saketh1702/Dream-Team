```python
import pandas as pd
```

```python
df = pd.read_csv('Data/ball_by_ball.csv')
df_player = pd.read_csv('Data/player.csv')
df_match = pd.read_csv('Data/match.csv')
df_player_match = pd.read_csv('Data/player_match.csv')
df_season = pd.read_csv('Data/season.csv')
df_team = pd.read_csv('Data/team.csv')
```

```python
class player_profile:
    def __init__(self,id):
        self.id = id
        self.detailed_scores = [0,0,0,0,0,0,0]
        self.score = []
        self.total_runs = 0
        self.match_ids = set(df_player_match[df_player_match['Player_Id'] == self.id]['Match_Id'])
        self.dismissal_types = list(df[(df['Bowler_Id']==self.id)]['Dissimal_Type'])
        self.match_wise_scores = [0]
        self.no_of_dissmals = 0
        self.no_of_wickets = 0
        self.number_of_overs_bowled = 0
        self.dismissal_types = []
        self.matchwise_wickets = [0]
        self.runsgiven_match_wise = [0]
        self.feldingtype = list(df[df['Fielder_Id'] == self.id]['Dissimal_Type'])
        self.runouts = 0
        self.catches = 0
        self.stumps = 0
        self.orangecap = 0
        self.purplecap = 0
        self.playerofseries = 0
        self.batting_average = 0.00
        self.bowling_average = 0.00
        self.strike_rate = 0.00
        self.economy_rate = 0.00
        self.numofballs = 0
        self.number_of_wickets = 0
        
        
    def get_scores(self):
        lis = list(df[df['Striker_Id'] == self.id]['Batsman_Scored'].values)
        self.score = [int(x) for x in lis if x not in [' ','Do_nothing'] ]
        self.total_runs = sum(self.score)
        for i in self.score:
            self.detailed_scores[i]+=1
    
    def get_number_of_fours(self):
        return(self.detailed_scores[4])
    
    def get_number_of_sixes(self):
        return(self.detailed_scores[6])
    
    def get_number_of_dot_balls(self):
        return(self.detailed_scores[0])
    
    def get_total_runs(self):
        return(self.total_runs)
    
    def get_total_num_of_balls(self):
        self.numofballs = sum(self.detailed_scores)
    
    def get_total_matches(self):
        return(len(self.match_ids))
    
    def get_match_wise_scores(self):
        for i in self.match_ids:
            s = list(df[(df['Striker_Id'] == self.id ) & (df['Match_Id']==i)]['Batsman_Scored'].values)
            s = [int(x) for x in s if x not in [' ','Do_nothing'] ]
            self.match_wise_scores.append(sum(s))
            
    def number_of_half_centuries(self):
        count = 0
        for i in self.match_wise_scores:
            if (i>=50) and (i<100):
                count+=1
        return(count)
    
    def number_of_centuries(self):
        count = 0
        for i in self.match_wise_scores:
            if (i>=100) and (i<200):
                count+=1
        return(count)
    
    def number_of_ducks(self):
        count = 0
        for i in self.match_wise_scores:
            if i == 0:
                count+=1
        return (count)
    
    def get_best_score(self):
        return(max(self.match_wise_scores))
    
    def  get_number_of_outs(self):
        li = list(df[df['Player_dissimal_Id']==str(self.id)]['Player_dissimal_Id'].values)
        li = [int(x) for x in li if x!=' ']
        self.no_of_dissmals = len(li)
        
    
    def get_number_of_wickets(self):
        count = 0
        for i in list(df[(df['Bowler_Id']==self.id)]['Dissimal_Type']):
            if i in['bowled','caught','caught and bowled','lbw','stumped']:
                count+=1
        return(count)
        
    def get_number_of_oversbowled(self):
        overs = []
        for i in self.match_ids:
            li = set(df[(df['Bowler_Id']==self.id) & (df['Match_Id']==i)]['Over_Id'])
            overs.append(len(li))
        self.number_of_overs_bowled = sum(overs)
        return(self.number_of_overs_bowled)
    
    def get_match_wise_wicket(self):
        for i in self.match_ids:
            li = list(df[(df['Bowler_Id']==self.id) & df['Match_Id']==i]['Dissimal_Type'])
            l = []
            for x in li :
                if x in ['bowled','caught','caught and bowled','lbw','stumped']:
                    l.append(x)
            self.matchwise_wickets.append(len(l))
    
    def highest_wickets(self):
        return(max(self.matchwise_wickets))
    
    def fifersinamatch(self):
        return(self.matchwise_wickets.count(5))
    
    def get_number_of_runs_given(self):
        for i in self.match_ids:
            li = list(df[(df['Bowler_Id']==self.id) & (df['Match_Id']==i)]['Batsman_Scored'])
            li = [int(x) for x in li if x not in [' ','Do_nothing']]
            self.runsgiven_match_wise.append(sum(li))
    
    def highest_runs_given(self):
        return(max(self.runsgiven_match_wise))
    
    
    def number_of_catches(self):
        feldingtype = list(df[df['Fielder_Id'] == self.id]['Dissimal_Type'])
        self.catches = feldingtype.count(['caught','caught and bowled'])
        return(self.catches)
    
    def number_of_runouts(self):
        self.runouts = self.feldingtype.count('run out')
        return(self.runouts)
    
    def number_of_stumps(self):
        self.stumps = self.feldingtype.count('stumped')
        return(self.stumps)
    
    def number_of_orangecaps(self):
        li = list(df_season['Orange_Cap_Id'])
        self.orangecap = li.count(self.id)
        return(self.orangecap)
    
    def number_of_purplecaps(self):
        li = list(df_season['Purple_Cap_Id'])
        self.purplecap = li.count(self.id)
        return(self.purplecap)
    
    def number_of_series(self):
        li = list(df_season['Man_of_the_Series_Id'])
        self.playerofseries = li.count(self.id)
        return(self.playerofseries)
    
    def battingaverage(self):
        if self.no_of_dissmals !=0:
            self.batting_average = (self.total_runs)/(self.no_of_dissmals)
        else:
            self.batting_average = 0
        return(self.batting_average)
    
    def bowlingaverage(self):
        self.bowling_average = (sum(self.runsgiven_match_wise))//(sum(self.matchwise_wickets))
    def strikerate(self):
        if self.numofballs !=0:
            self.strike_rate = (self.total_runs*100)/(self.numofballs)
        else:
            self.strike_rate = 0
        return(self.strike_rate)
    
    def economy(self):
        self.economy_rate = (sum(self.runsgiven_match_wise))//(self.number_of_overs_bowled)
        return(self.economy_rate)
                
```
